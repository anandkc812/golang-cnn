```python

#Extract accuracy - values from log file - into a list
def get_acc_from_file(file_name):
    
    file1 = open(file_name, 'r') 
    Lines = file1.readlines() 

    count = 0
    search_str = "Accuracy "
    search_len = len(search_str)

    result = []
    for line in Lines: 
        start = line.find(search_str)

        end   = len(line)
        if start != -1 :
            result.append(float(line[start+search_len:end]))
    
    return result
```

```python
import matplotlib.pyplot as plt

result1 = get_acc_from_file("new-log-lrelu-0001_july.logs")
result2 = get_acc_from_file("new-log-tanh-0001_july.logs")
result3 = get_acc_from_file("new-log-lrtanh-0001_july.logs")
result4 = get_acc_from_file("new-log-sigmoid-0001_july.logs")

plt.rcParams['figure.figsize'] = [20, 10]
plt.title(" Comparisions : lrelu vs tanh vs lrtanh, lr=0.0001",fontsize = 'x-large')
plt.xlabel("Epocs",fontsize = 'x-large')
plt.ylabel("Accuracy in %",fontsize = 'x-large')

plt.plot(result1)
plt.plot(result2)
plt.plot(result3)
plt.plot(result4)

 #["blue", "green","orange","red"],
    
plt.legend( ["lrelu", "tanh", "lrtanh", "sigmoid"],loc ="lower right",fontsize = 'x-large')
plt.grid()
plt.show()

```
